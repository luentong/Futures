```python
x = []
y = []
```

```python
import csv
file = open('ag/2101.csv')
csvreader = csv.reader(file)
rows = []
for row in csvreader:
        if row:
            rows.append(row)
```

```python
print(len(rows))
```

```python
import numpy
index = 0
for i in range(len(rows) - 10):
    x_add = []
    for j in range(10):
        for item in rows[i + j][2:]:
            print(item)
            if item == "0":
                x_add.append(0)
            else:
                x_add.append(float(item))
#     print(x_add)
#     print(len(x_add))
    x.append(numpy.array(x_add))
#     print(i)
#     print(rows[i+30])
#     print(rows[i+29])
#     print("wow")
    y_add = float(rows[i+10][5]) - float(rows[i+9][5])
    y.append(y_add)
    print(y_add)
```

```python
# 进行训练和测试样本的分割
from sklearn.model_selection import train_test_split
```

```python
# 90%的样本用来训练，剩余10%用来测试
x_ar = numpy.array(x)
y_ar = numpy.array(y)
X_train, X_test, y_train, y_test = train_test_split(x_ar,y_ar,
                                                    test_size=0.1,
                                                    random_state=12)
```

```python
# 用LSTM对样本进行分类
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import RMSprop
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping, ModelCheckpoint, TensorBoard, ReduceLROnPlateau
model = Sequential()
model.add(Dense(100, input_dim=100, kernel_initializer='normal', activation='linear'))
model.add(Dense(70, activation='linear'))
model.add(Dense(40, activation='linear'))
model.add(Dense(20, activation='linear'))
model.add(Dense(1, activation='linear'))
```

```python
# 我们使用adam以0.001的learning rate进行优化
optimizer = Adam(lr=1e-3)
```

```python
import tensorflow as tf
# model.compile(loss='mse',
#               optimizer=optimizer,
#               metrics=[tf.keras.metrics.MeanSquaredError()])

model.compile(loss='mse',
              optimizer=optimizer,
              metrics=['mse'])
```

```python
# 我们来看一下模型的结构，一共90k左右可训练的变量
model.summary()
```

```python
# 定义early stoping如果3个epoch内validation loss没有改善则停止训练
earlystopping = EarlyStopping(monitor='val_loss', patience=1000, verbose=1)
```

```python
# 自动降低learning rate
lr_reduction = ReduceLROnPlateau(monitor='val_loss',
                                       factor=0.1, min_lr=1e-8, patience=0,
                                       verbose=1)
```

```python
# 定义callback函数
callbacks = [
    earlystopping, 
    #lr_reduction
]
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
# 开始训练
model.fit(X_train, y_train,
        validation_split=0.1, 
          epochs=4000,
        batch_size=45,
    callbacks=callbacks)
```
